```python
from InstagramAPI import InstagramAPI
import json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Logging into Instagram API
* username and password is needed

```python
api = InstagramAPI("username", "userPassword")
api.login()
#you should get "Login success!" as response
```

## Use function from Instagram API to get result
* print(dir(InstagramAPI), to get the list of all the functions

```python
api.searchLocation('goa')
```

```python
#convert the result into JSON format
result = api.LastJson

with open('search_location.json', 'w') as outfile:
    outfile.write(json.dumps(result))
```

## Normalize the received JSON data
### The received result contains following keys, 
* items, has_more, inform_module, page_token, rank_token, status
* result['items'] has all the location based information

```python
db_json = json_normalize(result['items'])
db_json
```

## Convert JSON data to pandas DataFrame

```python
db_df = pd.DataFrame(db_json)
db_df.reset_index(drop=True, inplace=True)
db_df.shape
db_df
```

## Filtered the data to required columns only
* original data contains 55 rows and 12 columns
#### We need:
* location name
* location city
* location longitude
* location latitude

```python
db_clean = db_df[['location.name', 'location.city', 'location.lng', 'location.lat']]
db_clean
```

## Data Preprocessing

### 1] Check for NaN values
* drop the NaN values

```python
db_clean = db_clean.dropna()
db_clean.shape
```

### 2] Bring the floating point precision upto two decimal places
* This is needed if we are going KMeans clustering later in this project
* This may affect quality of the results produced

```python
db_clean[['location.lng', 'location.lat']] = db_clean[['location.lng', 'location.lat']].round(2)
```

### 3] Convert the longitude and latitude values of data into Numpy Array
* This is because we need only longitude and latitude values for KMeans Clustering 
* 1. reset the index of db_clean, because removing NaN omits the previous index
* 2. convert the longitude and latitude values into Numpy Array

```python
db_clean = db_clean.reset_index()
```

```python
db_clean.shape
```

```python
points = db_clean[['location.lng', 'location.lat']].to_numpy()
points.shape
```

### Just to verify the conversion was done properly without losing any values
* cross check the DF values and Numpy array

```python
for i, row in db_clean.iterrows():
    city1_name = row['location.name']
    city1_lng = row['location.lat']
    city1_lat = row['location.lng']
    
    print(city1_lat," ",city1_lng," ", points[i]," ",i)
```

## Create new DataFrame to store cluster information
* After applying we can store the labels of clusters in the last column

```python
db_final = db_clean
```

```python
db_final.columns
```

## Performing KMeans Clustering
* check the labels for the cluster results

```python
kmeans =KMeans(n_clusters=8, random_state=0).fit(points)
```

```python
kmeans.labels_
```

## Attach the results in final DataFrame

```python
#create new DF where all labels will be stored
labels = pd.DataFrame(kmeans.labels_)
labels.shape
```

```python
#create a series to insert random number
#later we can attach actual labels obtained from KMeans
lbl = pd.Series(np.random.randn(len(db_final['location.name'])))
```

```python
#attach the lable(lbl) column in the db_final
db_final = db_final.assign(lbl=lbl.values)
```

```python
#assign the actual labels(KMeans labels_) in db_final
db_final = db_final.assign(lbl=labels.values)
```
